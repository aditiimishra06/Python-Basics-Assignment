```python
# que 13  Write a Python program to find the sum of all even numbers between 1 and 50.
#ans.
even_sum = 0
for num in range (0,51):
  if num % 2 == 0:
    even_sum += num
print(f"the sum of even number from 1 to 50 is {even_sum}")
```

```python
#14. Write a program to reverse a string using a while loop.
#ans
string ="Pwskills"
rev_string =""
length =len(string)-1
while length >=0:
  rev_string += string[length]
  length -= 1
print(rev_string)

```
